```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the csv
df = pd.read_csv('Salary_Data.csv')
```

```python
#if data frame is having null values or anyt missing values
df.isna().sum()
```

```python
# find if experience and salary are related

covariance_value = np.cov(df['YearsExperience'],df['Salary'])
print(covariance_value)
```

```python
print(covariance_value[0][1])
```

```python
correlation_value= np.corrcoef(df['YearsExperience'],df['Salary'])
print(correlation_value[0][1])
```

```python
# DATA VISUALIZATION
```

```python
plt.plot(df['YearsExperience'],df['Salary'])
plt.xlabel("EXPERIENCE")
plt.ylabel("SALARY")
plt.title("YEARS OF EXP VS SALARY")
```
