```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the csv
df = pd.read_csv('50_Startups.csv')
```

```python
#if data frame is having null values or anyt missing values
df.isna().sum()
```

```python
# find if experience and salary are related

covariance_value1 = np.cov(df['RnD'],df['Profit'])
covariance_value2 = np.cov(df['Administration'],df['Profit'])
covariance_value3 = np.cov(df['Marketing'],df['Profit'])

print(f"RnD and Profit = {covariance_value1[0][1]}")
print(f"Administration and Profit = {covariance_value2[0][1]}")
print(f"Marketing and Profit = {covariance_value3[0][1]}")
```

```python
correlation_value1= np.corrcoef(df['RnD'],df['Profit'])
correlation_value2= np.corrcoef(df['Administration'],df['Profit'])
correlation_value3= np.corrcoef(df['Marketing'],df['Profit'])


print(f"RnD and Profit {correlation_value1[0][1]}")
print(f"Adminstration and Profit {correlation_value2[0][1]}")
print(f"Marketing and Profit {correlation_value3[0][1]}")
```

```python
plt.plot(df['RnD'],df['Profit'])
plt.xlabel("RnD")
plt.ylabel("Profit")
plt.title("RnD VS Profit")
```

```python
plt.plot(df['Administration'],df['Profit'])
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.title("Administration VS Profit")
```

```python
plt.plot(df['Marketing'],df['Profit'])
plt.xlabel("Marketing")
plt.ylabel("Profit")
plt.title("Marketing VS Profit")
plt.savefig("Marketting_Profit")
```
