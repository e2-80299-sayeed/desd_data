```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# get the data
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# remove the column
df = df.drop(['CustomerID','Gender','Age'],axis=1)
```

```python
df.head()
```

```python
#Kmeans
```

```python
from sklearn.cluster import KMeans

#model is created
model = KMeans(n_clusters=5)

# train the model
model.fit(df)
```

```python
print(model.labels_)
```

```python
print(model.cluster_centers_)
```

```python
# Visualization 
```

```python
plt.figure(figsize=(10,8))
colors = ['red','green','blue','black','orange']

for k in range(5):
    plt.scatter(
        df['Annual Income (k$)'][model.labels_==k],
        df['Spending Score (1-100)'][model.labels_==k],
        color = colors[k],
        s=40,
        label = f"Cluster {k}")
    
plt.legend()
```

```python
plt.figure(figsize=(10,8))
colors = ['red','green','blue','black','orange']
markers = ['.','^','1','s','d']

for k in range(5):
    plt.scatter(
        df['Annual Income (k$)'][model.labels_==k],
        df['Spending Score (1-100)'][model.labels_==k],
        color = colors[k],
        marker = markers[k],
        s=40,
        label = f"Cluster {k}")
    
plt.legend()
```

```python
#Visualization using sns
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df,x='Annual Income (k$)',y='Spending Score (1-100)',
               c = model.labels_)
```
